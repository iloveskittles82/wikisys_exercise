Daten und Vorbereitung

```python
# Importieren der benötigten Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Erstellen der Daten
X = np.array([30, 50, 70, 90, 110, 130]) # Geschwindigkeit in km/h
y = np.array([8.5, 7.8, 7.2, 7.0, 7.4, 8.3]) # Verbrauch in l/100 km

# Umwandeln der Daten in die richtige Form für scikit-learn
X = X.reshape(-1, 1) # Umwandeln von eindimensionalem Array in zweidimensionales Array
y = y.reshape(-1, 1) # Umwandeln von eindimensionalem Array in zweidimensionales Array
```

Zwei verschiedene Modelle erstellen

```python
# Erstellen und Anpassen eines linearen Regressionsmodells
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Erstellen und Anpassen eines polynomiellen Regressionsmodells vom Grad 2
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X) # Erzeugen der zusätzlichen Potenz von x
poly_lin_reg = LinearRegression()
_ = poly_lin_reg.fit(X_poly, y)
```

Welches Modell ist besser?

```python
# Vorhersagen mit beiden Modellen
y_pred_lin = lin_reg.predict(X) # Vorhersage mit linearem Modell
y_pred_poly = poly_lin_reg.predict(X_poly) # Vorhersage mit polynomiellem Modell

# Berechnen des Bestimmtheitsmaßes R² für beide Modelle
r2_lin = lin_reg.score(X, y) # R² für lineares Modell
r2_poly = poly_lin_reg.score(X_poly, y) # R² für polynomielles Modell

# Ausgeben der Koeffizienten und des Bestimmtheitsmaßes für beide Modelle
print("Lineares Modell: y = {:.3f} + {:.3f}x".format(lin_reg.intercept_[0], lin_reg.coef_[0][0]))
print("R² für lineares Modell: {:.3f}".format(r2_lin))
print("Polynomielles Modell: y = {:.3f} + {:.3f}x + {:.3f}x²".format(poly_lin_reg.intercept_[0], poly_lin_reg.coef_[0][1], poly_lin_reg.coef_[0][2]))
print("R² für polynomielles Modell: {:.3f}".format(r2_poly))

# Plotten der Daten und der Modelle
plt.scatter(X, y, color="blue", label="Daten")
plt.plot(X, y_pred_lin, color="red", label="Lineares Modell")
plt.plot(X, y_pred_poly, color="green", label="Polynomielles Modell")
plt.xlabel("Geschwindigkeit (km/h)")
plt.ylabel("Verbrauch (l/100 km)")
plt.legend()
plt.show()
```
